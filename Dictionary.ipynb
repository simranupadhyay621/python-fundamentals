```python
# Empty dictionary
d = {}

# With values
d = {"name": "Harish", "age": 25, "city": "Delhi"}
print(d)
```

```python
#Removes all items.
d = {"a": 1, "b": 2}
d.clear()
print(d)   # {}
```

```python
#Creates a shallow copy.
d1 = {"a": 1, "b": 2}
d2 = d1.copy()
print(d2)
```

```python
#Creates dictionary with given keys and same value.
keys = ["a", "b", "c"]
d = dict.fromkeys(keys, 0)
print(d)
# {'a': 0, 'b': 0, 'c': 0}
```

```python
#Returns value of key (no error if key missing).
d = {"name": "Harish", "age": 25}
print(d.get("name"))      # Harish
print(d.get("salary"))    # None
print(d.get("salary", 0)) # 0 (default value)
```

```python
#Returns all keys.
d = {"a": 1, "b": 2}
print(d.keys())

for k in d.keys():
    print(k)
```

```python
#Returns all Values.
d = {"a": 1, "b": 2}
print(d.values())

for v in d.values():
    print(v)
```

```python
#Removes specific key.
d = {"a": 1, "b": 2}
d.pop("a")
print(d)

# With default value
d.pop("x", "Not Found")
```

```python
#Removes last inserted item
d = {"a": 1, "b": 2}
d.popitem()
print(d)
```

```python
#Returns key-value pairs.
d = {"a": 1, "b": 2}
print(d.items())
```

```python
#Updates dictionary with another dictionary.
d = {"a": 1, "b": 2}
d.update({"b": 20, "c": 30})
print(d)
```

```python
#Add / Update value
d = {"a": 1}
d["b"] = 2
d["a"] = 10
print(d)
```

```python
#Delete using del
d = {"a": 1, "b": 2}
del d["a"]
print(d)
```

```python
#Check key exists
d = {"a": 1}
print("a" in d)  # True
```

```python
#Print length of dictionary
print(len(d))
```
