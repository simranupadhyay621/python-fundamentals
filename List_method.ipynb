```python
#list in python
marks=[98.7,67.8,77.9,54.8,85.8]
print(marks)
print(type(marks))
print(marks[0])
print(len(marks))
```

```python
#list slicing
marks=[87,98,67,77,98,59]
print(marks[0:4])
```

```python
#list method
list=[1,2,3,5,6]
print(list.append(8))
print(list)
```

```python
list=[3,7,4,6,5,8]
print(list.sort())
print(list)
```

```python
#reverse
list=[3,7,4,6,5,8]
print(list.reverse())     
print(list)
```

```python
#insert(inx,el)

list=[3,7,4,6,5,8]
print(list.insert(3,9))      #insert(inx,el)
print(list)
```

```python
list=[2,4,2,5,3,6,2,7,9]
print(list.remove(2))     
 #remove first occurrence
print(list)
```

```python
#remove idx

list=[2,4,2,5,3,6,2,7,9]
print(list.pop(3))      
print(list)
```
