```python
 # 1. Create a Tuple of  10 sports names, do all operations addition,remove, change elements in the Tuple. 
```

```python
sports_tup=("Swimming","Hockey","Football","Archery","Running","Baseball","Tennis","Badmintion","Shooting")
#convert to list (tupple is imutable)
sports_list=list(sports_tup)
print(sports_list)
```

```python
#Adding new element
sports_list.append("Swimming")
print(sports_list)
```

```python
#Removing element
sports_list.remove("Shooting")
print(sports_list)
```

```python
#pop element removes last element
sports_list.pop()
print(sports_list)
```

```python
#pop element by index
sports_list.pop(1)
print(sports_list)
```

```python
#Extending the list
sports_new=("Chess","Javelin Throw","Shooting")
sports_new=list(sports_new)
sports_list.extend(sports_new)
print(sports_list)
```

```python
#Change the element by index number
sports_list[5]="Table Tennis"
print(sports_list)                    #Replaces tennis by Table Tennis
```

```python
#Get the Position of element by Index number
print(sports_list.index("Baseball"))
```

```python
#Insert new sports at specific position
sports_list.insert(1,"Long Jump")
print(sports_list)
```

```python
#Sort the element by ascending order
sports_list.sort()
print(sports_list)
```

```python
#Sort the element by descending order
sports_list.sort(reverse=True)
print(sports_list)
```

```python
#Copy list to another list
sports=sports_list.copy()
print(sports)
```

```python
#Reverse the list
sports.reverse()
print(sports)
```

```python
#convert list to tuple
sports=tuple(sports)
print(sports)
```

```python
# 2. Write a python program to interchange a list items with & without using temporary values.
```

```python
#Without Temporary method
list=["A","B","C","D","E","F","G"]
pos1=1
pos2=3
list[pos1],list[pos2]=list[pos2],list[pos1]
print("After Interchanging list:",list)
```

```python
# using Temporary value
idx_1 = 1
idx_2 = 4
print("List Before interchanging:",list)
temp = list[idx_1]
list[idx_1] = list[idx_2]
list[idx_2] = temp
print("After Interchanging with temp:",list)
```

```python
# 3. Create a dictionary, do all operations like addition of elements, remove elements, copy items & create nested dictionaries too.
```

```python
#Create a dictionary
patient = {
    "id": "P1023",
    "name": "Ravi Kumar",
    "age": 45,
    "diagnosis": "Hypertension",
}
print(patient)
```

```python
#Updating existing key value
patient["age"]=46
print(patient)
```

```python
#accessing elements
print(patient["name"])
print(patient.get("diagnosis"))
```

```python
#Removing Item using pop() method
patient.pop("diagnosis")
print(patient)
```

```python
#popitem() method removes the last inserted item
patient.popitem()
print(patient)
```

```python
#Removing Item using del method
del patient["name"]
print(patient)
```

```python
#merge the dictionaries
doctor = {
    "id": "D2045",
    "name": "Dr. Anil Sharma",
    "specialization": "Cardiology",
}
patient.update(doctor)
print(patient)
```

```python
#Return the keys of dictionary
print(patient.keys())
```

```python
#Return the values of dictionary
print(patient.values())
```

```python
#Return the items of dictionary in key value pair
print(patient.items())
```

```python
#Create nested dictionary
patient = {
    "name": "Ravi",
    "age": 45,
    "contact": {
        "phone": "9876543210",
        "email": "ravi@example.com"
    }
}
print(patient)
```

```python
#Accessing nested values
print(patient["contact"]["email"])
```

```python
#Adding element to nested dictionary
patient["contact"]["address"]="123, Main Street, City"
print(patient)
```

```python
#Update the nested dictionary
patient["contact"]["phone"]="9123456780"
print(patient)
```
