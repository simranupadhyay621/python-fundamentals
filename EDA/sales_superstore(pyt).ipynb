DATA VISUALISATION :

relplot:

```python
sns.relplot(df,x="Postal Code",y="Sales",hue="Ship Mode",kind="scatter")
plt.title("Postal_code Vs Sales by Ship_mode",color="r")
plt.show()
```

scatterplot:

```python
sns.scatterplot(df,x="Postal Code",y="Sales",hue="Region")
plt.title("SALES VS POSTAL_CODE")
plt.show()
```

Histplot:

```python
sns.histplot(df,x="Sub-Category",bins=15,kde=True)
plt.title("COUNT BY SUB-CATEGORY")
plt.xticks(rotation=90)
plt.show()
```

Boxplot:

```python
sns.boxplot(df,x="Region",y="Sales")
plt.title("POSTAL_CODE BY REGION",color="b")
plt.show()
```

voilinplot:

```python
sns.violinplot(df,x="Category",y="Sales",split=True)
plt.title("SALES BY CATEGORY")
plt.show()
```

barplot:

```python
plt.figure(figsize=(20,5))
sns.barplot(df,x="State",y="Sales")
plt.title("STATE WISE TOTAL SALES",color="r")
plt.xticks(rotation=90)
plt.show()
```

```python
top5_states = (
    df.groupby("State", as_index=False)["Sales"]
      .sum()
      .sort_values(by="Sales", ascending=False)
      .head(5)
)
sns.barplot(data=top5_states, x="State", y="Sales")
plt.title("TOP 5 STATES BY TOTAL SALES", color="r")
plt.show()
```

pairplot:

```python
sns.pairplot(df, hue='Region')
plt.show()
```

countplot:

```python
sns.countplot(data=df,x="Category")
plt.title("COUNT OF CATEGORY")
plt.show()
```
