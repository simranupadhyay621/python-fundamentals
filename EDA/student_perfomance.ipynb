#DATA VISUALISATION:

countplot:

```python
sns.countplot(sp,x="lunch",hue="gender")
plt.title("count of lunch",color="b")
plt.show()
```

barplot:

```python
plt.figure(figsize=(9, 6)) 
sns.barplot(sp,x="parental level of education",y="math score")
plt.title("LEVEL OF EDUCATION BY MATH_SCORE",color="y")
plt.show()
```

scatterplot:

```python
sns.scatterplot(sp,x="math score",y="reading score",hue="gender")
plt.title("MATH_SCORE VS READING_SCORE",color="k")
plt.show()
```

histplot:

```python
plt.Figure(figsize=(15,10))
sns.histplot(sp,x="parental level of education",kde=True)
plt.title("TOP LEVEL OF EDUCATION",color="b")
plt.show()
```

distplot:

```python
sns.displot(sp,x="writing score",col="race/ethnicity")
plt.title("WRITING_SCORE BY RACE/ETHNICALLY",color="r")
plt.show()
```

lineplot:

```python
sns.lineplot(sp,x="reading score",y="writing score",hue="gender")
plt.title("GENDER WISE SCORE",color="b")
plt.show()
```

boxplot:

```python
sns.boxplot(sp,x="lunch",y="writing score")
plt.title("WRITING_SCORE BY LUNCH",color="g")
plt.show()
```

pairplot:

```python
sns.pairplot(sp, hue='gender')
plt.show()
```

relplot:

```python
sns.relplot(sp,x="math score",y="reading score",hue="test preparation course")
plt.title("SCORE OF MATH & READING BY PREPARATION_COURSE",color="y")
plt.show()
```

voilinplot:

```python
sns.violinplot(sp,x="test preparation course",y="writing score",split=True)
plt.title("WRITING_SCORE BY TEST PREPARATION",color="g")
plt.show()
```
