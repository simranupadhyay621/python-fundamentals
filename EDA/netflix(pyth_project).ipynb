DATA VISUALISATION

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='rating', order=df['rating'].value_counts().index)
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title("Top 10 Content Producing Countries")
plt.show()
```

```python
movies = df[df['type'] == 'Movie']

sns.histplot(movies['duration'], bins=30, kde=True)
plt.title("Movie Duration Distribution",color="b")
plt.show()
```

```python
sns.scatterplot(
    data=movies,
    x='release_year',
    y='duration'
)
plt.title("Movie Duration vs Release Year",color="red")
plt.show()

```

```python
sns.boxplot(x=movies['duration'], color='skyblue')
plt.title("Boxplot of Movie Duration (Minutes)")
plt.show()
```

```python
sns.countplot(df,x="rating")
plt.title("rating wise movies count",color="y")
plt.show()

```
