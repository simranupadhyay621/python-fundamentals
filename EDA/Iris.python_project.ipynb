# EXPLORATORY DATA ANALYSIS
  

```python
plt.figure(figsize=(5,3))
a=sns.countplot(data=Ir,x="species")
plt.bar_label(a.containers[0])
plt.title("count by species",color="k")
plt.show
```

```python
plt.figure(figsize=(4,4))
pie_data=Ir.groupby("species")["petal_length"].sum()
plt.pie(pie_data,autopct="%1.1f%%")
plt.title("petal_length distribution by species")
plt.show
```

```python
sns.lineplot(Ir,x="sepal_length",y="petal_length",hue="species")
plt.title("sepal_length vs petal_length by species",color="b")
plt.show
```

```python
sns.relplot(Ir,x="sepal_width",y="petal_width",hue="species",kind="scatter")
plt.title("sepal_width vs petal_width by species",color="purple")
plt.show
```

```python
sns.displot(Ir,x="petal_width",col="species",kde=True)
plt.show
```

```python
sns.boxplot(Ir,x="species",y="sepal_length")
plt.title("sepal_length by species",color="k")
```

```python
sns.histplot(Ir,x="sepal_length",bins=15,kde=True)
plt.title("count by sepal_length")
plt.show
```

```python
sns.violinplot(Ir,x="species",y="petal_width",split=True)
plt.title("petal_width by species")
plt.show()
```

```python
sns.scatterplot(Ir,x="species",y="petal_length")
plt.title("petal_length by species")
plt.show()
```

```python
corr=Ir.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="Set3")
plt.show()
```

```python
sns.pairplot(Ir[["sepal_length","sepal_width","petal_length","petal_width","species"]],hue="species")
plt.show()
```
