#DATA VISUALISATION

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Total Bill vs Tip')
plt.show()
```

```python
sns.boxplot(x='sex', y='tip', data=tips)
plt.title('Tip by Sex')
plt.show()
```

```python
sns.boxplot(x='smoker', y='tip', data=tips)
plt.title('Tip by Smoker')
plt.show()
```

```python
sns.barplot(x='day', y='tip', hue='sex', data=tips)
plt.title('Average Tip by Day and Sex')
plt.show()
```

```python
sns.pairplot(tips, hue='sex')
plt.show()
```

```python
corr = tips[['total_bill', 'tip', 'size']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.histplot(tips,x="total_bill",kde=True)
plt.show()
```

```python
sns.countplot(data=tips,x="time")
```

```python
sns.countplot(data=tips,x="time",hue="smoker")
```

```python
sns.lineplot(tips,x="size",y="total_bill",hue="smoker")
```

```python
sns.relplot(tips,x="tip",y="total_bill",hue="day",kind="scatter")
plt.title("tip vs total_bill by day ",color="purple")
plt.show
```

```python
sns.violinplot(tips,x="day",y="total_bill",hue="time",split=True)
plt.title("total_bill by day")
plt.show()
```

```python
sns.displot(tips,x="tip",col="sex",hue="smoker",kde=True,multiple="stack")
```
