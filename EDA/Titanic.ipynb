Statistical & Mathematical Methods

```python
a=titanic["age"].mean()
b=titanic["age"].median()
c=titanic["age"].mode()
d=titanic["fare"].sum()
e=titanic["fare"].max()
f=titanic["fare"].min()
g=titanic["fare"].std()
h=titanic["fare"].var()
print(f"Mean={a}\nMedian={b}\nMode={c}\nTotal Fare={d}\nMax Fare={e}\nMin fare={f}\nstd_dev={g}\nVariance={h}")
```

```python
survivors=titanic[titanic["survived"]==1]                   # Filter passengers who survived
death=titanic[titanic["survived"]==0]                       # Filter passengers who did not survive
total_survivars=survivors["survived"].count()               # Count survivors
total_death=death["survived"].count()                       # Count deaths
total_passengers=titanic["survived"].count()                # Total passengers

print(f"Total Survivors={total_survivars}, Mean age of survivors={survivors["age"].mean()}")
print(f"Total Death={total_death}, Mean age of death={death["age"].mean()}")
print(f"Total Passengers={total_passengers}")
print(f"Survival Ratio={round((total_survivars/total_passengers)*100,0)}%")
print(f"Death Ratio={round((total_death/total_passengers)*100,0)}%")
```

```python
#Survival Count (Count Plot)
```

```python
# Plot count of survived vs not survived
a=sns.countplot(x="survived", data=titanic)

# Add value labels on bars
plt.bar_label(a.containers[0])
plt.show()

#The plot shows that more passengers did not survive than survived.
```

```python
# Comparison of survival between male and female
a=sns.countplot(data=titanic,x="sex", hue="survived")
plt.show()
#Females had a much higher survival rate than males. This supports the historical rule of “women and children first.”
```

```python
# Histogram showing age distribution
a=sns.histplot(titanic,x="age", bins=30, kde=True)
plt.bar_label(a.containers[0])
plt.show()

#Most passengers were between 20 and 40 years old. There were fewer children and elderly passengers onboard.
```

```python
# Histogram comparing age distribution for survived vs not survived
a=sns.histplot(data=titanic, x="age", hue="survived", bins=30, kde=True, multiple="stack")
plt.bar_label(a.containers[0])
plt.bar_label(a.containers[1])
plt.show()

#Younger passengers, especially children, show relatively higher survival counts compared to older age groups.
```

```python
# Compute correlation matrix (numeric columns only)
corr = titanic.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="Set1")
plt.show()

"""Survival is positively correlated with fare and negatively correlated with pclass (lower class number = higher status).
This reinforces the importance of class and wealth."""
```

```python
# Pairwise relationships between numerical features
a=sns.pairplot(titanic[["age","fare","sibsp","parch","survived"]], hue="survived")
plt.show()

"""Survivors tend to cluster around higher fare values.
Pairwise plots also show separation between survived and non‑survived passengers for some features."""
```

```python
# Total fare collected by embarked
y=titanic.groupby("embarked")["fare"].sum().reset_index()
a=sns.barplot(data=y,x="embarked",y="fare")
plt.bar_label(a.containers[0])
plt.title("Fare by Embarked")
plt.show()

#Passengers embarking from S paid higher fares on average, indicating that wealthier passengers boarded from this port.
```

```python
# Sum survived by embarked
survived_sum = titanic.groupby("embarked")["survived"].sum()
# Create labels with values
labels = [f"{embark}\n{value}" for embark, value in zip(survived_sum.index, survived_sum.values)]

# Pie chart
plt.pie(survived_sum,labels=labels,autopct='%1.1f%%')
plt.title("Survival Distribution by Embarked")
plt.show()

#Most survivors embarked from Southampton (S) due to higher passenger volume.
```

```python
# Identify categorical columns
cat_cols = titanic.select_dtypes(include="object").columns
cat_cols
for col in cat_cols:
    plt.figure(figsize=(4,2))
    sns.countplot(y=col, data=titanic) 
    plt.title(f"Countplot of {col}")
    plt.show()
```

```python
plt.figure(figsize=(5,3))
a=sns.countplot(x="sex", hue="survived", data=titanic)
plt.title("Survival by Gender")
plt.bar_label(a.containers[0])
plt.bar_label(a.containers[1])
plt.show()

# Crosstab for numerical comparison
pd.crosstab(titanic['sex'], titanic['survived'], normalize='index')

#The numerical table confirms that females survived at a much higher proportion than males, validating the visual observation.
```

```python
plt.figure(figsize=(5,3))
a=sns.countplot(x="pclass", hue="survived", data=titanic)
plt.title("Survival by Passenger Class")
plt.bar_label(a.containers[0])
plt.bar_label(a.containers[1])
plt.show()
#Survival decreases as passenger class moves from 1st to 3rd, reinforcing the class‑based survival pattern.
```

```python
# count plot by survival status
g=sns.catplot(data=titanic,x="pclass",col="survived",kind="count",hue="sex")
for ax in g.axes.flat:
    for container in ax.containers:
        ax.bar_label(container)
plt.show()
#Females in 1st and 2nd class had the highest survival rates. Males in 3rd class were the most vulnerable group.
```

```python
# Create a new feature: family size
titanic['family_size'] = titanic['sibsp'] + titanic['parch'] + 1
a=sns.countplot(x="family_size", hue="survived", data=titanic)
plt.title("Family Size vs Survival")
plt.bar_label(a.containers[0])
plt.bar_label(a.containers[1])
plt.show()
#Passengers with small families (2–4 members) had better survival chances, while solo travelers and very large families had lower survival rates.
```
