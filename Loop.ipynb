```python
#Print Prime Numbers in a Given Range
start = int(input("Enter start number: "))
end = int(input("Enter end number: "))

for num in range(start, end + 1):
    if num > 1:
        for i in range(2, num):
            if num % i == 0:
                break  # Not prime
        else:
            print(num, "is Prime")
```

```python
#Factorial of a number
num = int(input("Enter a number: "))
fact = 1

for i in range(1, num + 1):
    fact *= i   # Multiply each number

print("Factorial:", fact)
```

```python
#Fibonacci Series.
n = int(input("Enter number of terms: "))

a, b = 0, 1

for i in range(n):
    print(a, end=" ")
    a, b = b, a + b   # Update values
```

```python
#Count Vowels in a String
text = input("Enter a string: ")
count = 0

for ch in text:
    if ch.lower() in "aeiou":
        count += 1

print("Total vowels:", count)
```

```python
#Reverse a Number
num = int(input("Enter a number: "))
reverse = 0

for i in range(len(str(num))):
    digit = num % 10
    reverse = reverse * 10 + digit
    num //= 10

print("Reversed number:", reverse)
```

```python
#Check Armstrong Number
num = int(input("Enter a number: "))
original = num
digits = len(str(num))
total = 0

for i in range(digits):
    digit = num % 10
    total += digit ** digits
    num //= 10

if total == original:
    print("Armstrong Number")
else:
    print("Not Armstrong")
```

```python
#Pattern Printing (Pyramid)
rows = int(input("Enter number of rows: "))

for i in range(1, rows + 1):
    print("*" * i)   # Print stars increasing
```

```python
#Inverted Right Triangle
rows = 5

for i in range(rows, 0, -1):
    print("*" * i)
```

```python
#Full Pyramid
rows = 5

for i in range(1, rows + 1):
    print(" " * (rows - i) + "* " * i)
```

```python
#Inverted full pyramid
rows = 5

for i in range(rows, 0, -1):
    print(" " * (rows - i) + "* " * i)
```

```python
#Diamond Pattern
rows = 5

# Upper half
for i in range(1, rows + 1):
    print(" " * (rows - i) + "* " * i)

# Lower half
for i in range(rows - 1, 0, -1):
    print(" " * (rows - i) + "* " * i)
```

```python
#Frequency of Each Character in a String
text = input("Enter a string: ")
freq = {}

for ch in text:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1

for key in freq:
    print(key, ":", freq[key])
```

```python
#Reverse a number using while loop
num = int(input("Enter a number: "))
reverse = 0

# Loop runs until number becomes 0
while num > 0:
    digit = num % 10          # Get last digit
    reverse = reverse * 10 + digit
    num //= 10                # Remove last digit

print("Reversed number:", reverse)
```

```python
#Check Palindrome
num = int(input("Enter a number: "))
original = num
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num //= 10

# Compare original and reversed number
if original == reverse:
    print("Palindrome Number")
else:
    print("Not a Palindrome")
```

```python
#Count Digits in a Number
num = int(input("Enter a number: "))
count = 0

# Special case for 0
if num == 0:
    count = 1
else:
    while num > 0:
        num //= 10
        count += 1

print("Total digits:", count)
```

```python
#ATM PIN Validation (3 Attempts)
correct_pin = "1234"
attempts = 0

while attempts < 3:
    pin = input("Enter PIN: ")

    if pin == correct_pin:
        print("Access Granted")
        break           # Exit loop if correct
    else:
        print("Wrong PIN")
        attempts += 1

if attempts == 3:
    print("Account Locked")
```

```python
#Find Sum of Digits
num = int(input("Enter a number: "))
total = 0

while num > 0:
    digit = num % 10
    total += digit
    num //= 10

print("Sum of digits:", total)
```
