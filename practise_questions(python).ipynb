```python
#Write a program that asks the user for a letter and prints if it's a vowel (a,e, i, o, u) or a consonant
 

s=input("enter a character")
if s in ("a","e","i","o","u","A","E","I","O","U"):
    print(f"{s} is an vowel")
else:
    print(f"{s} is consonant")
```
