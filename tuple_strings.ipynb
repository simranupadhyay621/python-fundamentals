```python
#create tuple in integer
tup=(1,2,3)
print(tup)
```

```python
#create tuple in string
tup_str=("a","b","c")
print(tup_str)
```

```python
#Mixed tuple
tup_mix=("a",1,3.4,True)
print(tup_mix)
print(type(tup_mix))
```

```python
#Single element will print the data type of tuple
x=("a")
print(type(x))
```

```python
#Reverse order
tup_mix=("a",1,3.4,True)
print(tup_mix[::-1])
```

```python
#Changing tuple to list to change the element then convert to list
tup_mix=("a",1,3.4,True)
tup_list=list(tup_mix)
tup_list[2]="xyz"
tup_list=tuple(tup_list)
print("Original List",tup_mix)
print("After changing tuple to list",tup_list)
```

```python
#Adding element to tuple
tup_mix=("a",1,3.4,True)
tup_list=list(tup_mix)   #converting to list
tup_list[2]="xyz"        #changing value
tup_list.append("new")   #Adding new value
tup_list=tuple(tup_list)
print("Original List",tup_mix)
print("After Adding new element",tup_list)
```
